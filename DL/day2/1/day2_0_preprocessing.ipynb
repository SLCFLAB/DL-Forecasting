```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_gold = pd.read_csv('Gold Futures Historical Data.csv',thousands = ',')
data_gold
```

```python
data_gold = data_gold[['Date','Price']]
data_gold = data_gold.astype({'Price':'float64'})
data_gold['Date'] = pd.to_datetime(data_gold['Date'], format='%m/%d/%Y')
data_gold.sort_values(by='Date', inplace=True)
data_gold
```

```python
data_silver = pd.read_csv('Silver Futures Historical Data.csv',thousands = ',')
data_silver = data_silver[['Date','Price']]
data_silver = data_silver.astype({'Price':'float64'})
data_silver['Date'] = pd.to_datetime(data_silver['Date'], format='%m/%d/%Y')
data_silver.sort_values(by='Date', inplace=True)
data_silver
```

```python
data_platinum = pd.read_csv('Platinum Futures Historical Data.csv',thousands = ',')
data_platinum = data_platinum[['Date','Price']]
data_platinum = data_platinum.astype({'Price':'float64'})
data_platinum['Date'] = pd.to_datetime(data_platinum['Date'], format='%m/%d/%Y')
data_platinum.sort_values(by='Date', inplace=True)
data_platinum
```

```python
df=pd.merge(data_gold, data_silver, on='Date', how='outer')
df
```

```python
df[df['Price_y'].isna()]
```

```python
df.fillna(method='ffill',inplace=True)
df
```

```python
data_gold
```

```python
data_platinum.reset_index(inplace=True,drop=True)
data_platinum.reset_index(inplace=True)
data_platinum.drop(columns=['Date'],inplace=True)
data_platinum.columns = ['ds','y']
data_platinum['unique_id'] = 'platinum'
mean_platinum = data_platinum['y'].mean()
std_platinum = data_platinum['y'].std()
data_platinum
```

```python
data_gold.reset_index(inplace=True,drop=True)
data_gold.reset_index(inplace=True)
data_gold.drop(columns=['Date'],inplace=True)
data_gold.columns = ['ds','y']
data_gold['unique_id'] = 'gold'
data_gold
```

```python
data_silver = df[['Date','Price_y']]
data_silver
```

```python
data_silver.reset_index(inplace=True,drop=True)
data_silver.reset_index(inplace=True)
data_silver.drop(columns=['Date'],inplace=True)
data_silver.columns = ['ds','y']
data_silver['unique_id'] = 'silver'
mean_silver = data_silver['y'].mean()
std_silver = data_silver['y'].std()
data_silver
```

```python
data_gold.to_csv('gold.csv',index=False)
data_silver.to_csv('silver.csv',index=False)
data_platinum.to_csv('platinum.csv',index=False)
```
